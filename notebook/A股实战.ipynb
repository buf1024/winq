```python
from datetime import datetime
from winq import *
import hiq_pyfetch as fetch
from winq.selector import *
db = default(log_level='info')

# test_end_date = '20230322'
# with_rt_kdata = False
run_task_count = 50

test_end_date = None
with_rt_kdata = True
```

```python
# 选择热门股票，热门股票必定属于某个热门的概念
strategy = strategies['test.RightSide'](db=db, 
                                        test_end_date=test_end_date, with_rt_kdata = with_rt_kdata,
                                        run_task_count=run_task_count,
                                        load_daily=db.load_stock_concept_daily, 
                                        load_info=db.load_stock_concept,
                                        fetch_daily=fetch.fetch_stock_concept_daily
                                       )
await strategy.prepare(min_hit_days=1,
                       max_down_in_rise=1.0,
                       max_up_in_rise=10.0,
                       max_falling_down=1.0,
                       max_turnover=10.0,
                       min_high_ndays=2,
                       test_recent_ndays=1,
                       test_recent_ndays_up=1.0,
                       sort_by='hit_chg_pct')

data = await strategy.run()
# 观察热门概念，可能需要微调参数

if data is not None and  len(data) > 0:
    data = data[~data['name'].str.contains('昨日')].reset_index(drop=True)
    test_end_date = strategy.test_end_date.strftime('%Y%m%d')
    export_codes_to_path(path='/Users/luoguochun/Downloads/stock/trade/hot_concept_{}.txt'.format(test_end_date), data=data)
data
```

```python
data_detail = await db.load_stock_concept_detail(filter={'code': {'$in': list(data['code'].values)}})
codes = list(set(data_detail['stock_code'].values))
len(codes)
```

```python
# codes=None
strategy = strategies['test.RightSide'](db=db, 
                                        test_end_date=test_end_date,
                                        with_rt_kdata = with_rt_kdata,
                                        run_task_count=run_task_count,
                                        load_daily=db.load_stock_daily, 
                                        load_info=db.load_stock_info,
                                        fetch_daily=fetch.fetch_stock_bar
                                       )
await strategy.prepare(min_hit_days=1,
                       max_down_in_rise=3.0,
                       max_up_in_rise=20.0,
                       max_falling_down=2.0,
                       max_turnover=10.0,
                       min_high_ndays=3,
                       test_recent_ndays=3,
                       test_recent_ndays_up=3.0,
                       sort_by='hit_chg_pct')

data = await strategy.run(codes=codes)
if data is not None and len(data) > 0:
    test_end_date = strategy.test_end_date.strftime('%Y%m%d')
    export_codes_to_path(path='/Users/luoguochun/Downloads/stock/trade/hot_stock_{}.txt'.format(test_end_date), data=data)

data
```

```python
path='/Users/luoguochun/Downloads/stock/analyze/supper_falling_20230329.txt'
codes=None
with open(path) as f:
    codes = f.read()
    codes = codes.split(',')
    codes = ['sh'+code if code.startswith('6') else 'sz'+code for code in codes]


# 热门股票超跌
strategy = strategies['test.SuperFalling'](db=db, 
                                        test_end_date=test_end_date, with_rt_kdata = with_rt_kdata,
                                        run_task_count=run_task_count,
                                        load_daily=db.load_stock_daily, 
                                        load_info=db.load_stock_info,
                                        fetch_daily=fetch.fetch_stock_bar
                                       )
await strategy.prepare(is_trading=True,
                        cal_time='10:30:00',
                        max_down_as_rise=-1.0,
                        max_20cm_rise=5.0,
                        max_20cm_falling=2.0,
                        max_10cm_rise=3.0,
                        max_10cm_falling=1.50,
                        sort_by='chg_pct')

data = await strategy.run(codes=codes)

if data is not None and  len(data) > 0:
    test_end_date = strategy.test_end_date.strftime('%Y%m%d')
    export_codes_to_path(path='/Users/luoguochun/Downloads/stock/trade/supper_falling_{}.txt'.format(test_end_date), data=data)
data
```
